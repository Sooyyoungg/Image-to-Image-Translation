```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

data_info = pd.read_csv('/share/home/connectome/conmaster/GANBERT/abcd_t1_t2_diffusion_info_final.csv')
```

# ABCD Sites 
### : show how many subjects are allocated in each abcd site
### values : Site ID at each event

```python
data_info.abcd_site.value_counts().sort_values(ascending=False).plot.bar()
```

# Mean pixel intensity
### : show differences of mean pixel intensity between abcd sites

```python
### show mean pixel intensity of t1 images according to abcd sites
data_info[['t1_mean']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
```

```python
### show mean pixel intensity of t2 images according to abcd sites
data_info[['t2_mean']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
```

```python
### show mean pixel intensity of fa images according to abcd sites
data_info[['fa_mean']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
```

```python
### show mean pixel intensity of md images according to abcd sites
data_info[['md_mean']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
```

```python
### show mean pixel intensity of dwi images according to abcd sites
data_info[['dwi_mean']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
```

# Sex 
### : show how many females and males are in data
### values : 1 = Female; 0 = Male

```python
data_info.sex.value_counts().sort_values(ascending=False).plot.bar()
```

```python
plt.figure(figsize=(20, 21))

for i in range(21):
    plt.subplot(1, 21, i+1)
    data_info[data_info['abcd_site']==i+1]['sex'].hist()
    plt.title('abcd_site: '+str(i))
```

```python
data_info[['sex']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
plt.xticks(rotation=350)
plt.show()
```

## => there are some sites where the sex ratio of subjects is quite different BUT, no big differences in Sex in terms of ABCD sites

# Age 
### : show how old the subjects are
### values : Age in months at the time of the interview/test/sampling/imaging

```python
data_info.age.value_counts().sort_values(ascending=False).plot.bar()
```

```python
data_info[['age']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
plt.xticks(rotation=350)
plt.show()
```

## => there are some ages that many subjects are allocated in BUT, no big differences in Age in terms of ABCD sites

# RACE 
### : show which race the subjects are allocated in
### values : 1 = White; 2 = Black; 3 = Hispanic; 4 = Asian; 5 = Other

```python
data_info.loc[:, 'race.ethnicity'].value_counts().sort_values(ascending=False).plot.bar()
```

```python
data_info[['race.ethnicity']].groupby(data_info['abcd_site']).mean().plot(kind='bar')
plt.xticks(rotation=350)
plt.show()
```

## => there are differences in the proportion of races in terms of sites
